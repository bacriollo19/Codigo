```python
import cv2
import os
import random
```

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        images.append(filename)
    return images
```

```python
frames = load_images_from_folder("images/Frame10segJPG")
print(len(frames))
f1=random.sample(frames,int(len(frames)*0.7))
print(len(f1))
f2=list(set(frames)-set(f1))
print(len(f2))
```

```python
import shutil
```

```python
def load_images(folder):
    for filename in os.listdir(folder):
        if filename in f1:
            shutil.copy(folder+filename, "images/train10/"+filename)
        elif filename in f2:
            shutil.copy(folder+filename, "images/test10/"+filename)
```

```python
load_images('images/Frame10segJPG/')
```

```python
def load_xml(folder):
    for filename in os.listdir(folder):
        if filename[0:len(filename)-3]+'jpg' in f1:
            shutil.copy(folder+filename, "images/xml3/"+filename)
        elif filename[0:len(filename)-3]+'jpg' in f2:
            shutil.copy(folder+filename, "images/xml4/"+filename)
```

```python
load_xml('images/Frame10segXML/')
```
